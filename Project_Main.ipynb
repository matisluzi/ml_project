```python
import pandas as pd
from datetime import timedelta
# Federal Reserve CSV files
fed_files = ["MORTGAGE30US.csv", "RRVRUSQ156N.csv", "CPIAUCSL.csv"]

# Read and concatenate Federal Reserve data
dfs_fed = [pd.read_csv(f, parse_dates=True, index_col=0) for f in fed_files]
fed_data = pd.concat(dfs_fed, axis=1)
fed_data.ffill(inplace = True)
```

```python
# Zillow CSV files
zillow_files = [
    "Metro_zhvi_bdrmcnt_1_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv",
    "Metro_zhvi_bdrmcnt_2_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv",
    "Metro_zhvi_bdrmcnt_3_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv",
    "Metro_zhvi_bdrmcnt_4_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv",
    "Metro_zhvi_bdrmcnt_5_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv"
]

# Read Zillow CSV files into a list of DataFrames
dfs_zillow = [pd.read_csv(f, parse_dates=True, index_col=0) for f in zillow_files]
dfs_zillow[0]
```

```python
# Debugging: Print length of dfs_zillow
print("Length of dfs_zillow:", len(dfs_zillow))
```

```python
# Example: Selecting a specific Zillow DataFrame based on input
LOCATION_INPUT = 0
BEDROOM_INPUT = 1

# Debugging: Print BEDROOM_INPUT and LOCATION_INPUT
print("BEDROOM_INPUT:", BEDROOM_INPUT)
print("LOCATION_INPUT:", LOCATION_INPUT)

# Ensure BEDROOM_INPUT and LOCATION_INPUT are within valid range
if 1 <= BEDROOM_INPUT < len(dfs_zillow) and 0 <= LOCATION_INPUT < len(dfs_zillow[BEDROOM_INPUT]):
    zillowFileData = dfs_zillow[BEDROOM_INPUT-1].iloc[LOCATION_INPUT, 5:]  # Extract relevant data

    # Create DataFrame from extracted data
    zillow_data_df = pd.DataFrame(zillowFileData)

    # Optional: Clean up DataFrame (e.g., handle missing values)
    zillow_data_df = zillow_data_df.dropna()  # Drop rows with NaN values if needed

    # Now you can work with zillow_data_df as your processed Zillow data
    print(zillow_data_df.head())  # Example: Display first few rows of processed data
else:
    print("Invalid BEDROOM_INPUT or LOCATION_INPUT. Check input values.")
zillow_data_df.index = pd.to_datetime(zillow_data_df.index)
zillow_data_df["month"] =  zillow_data_df.index.to_period("M")
```

```python
fed_data.index = fed_data.index + timedelta(days=2)
price_data = zillow_data_df.merge(fed_data, how='left', left_index=True, right_index=True)
```

```python
price_data.columns = ["price","interest", "vacancy", "cpi"]
price_data['interest']=price_data['interest'].ffill()
price_data['vacancy']=price_data['vacancy'].ffill()
price_data['cpi']=price_data['cpi'].ffill()
```

```python
price_data.plot.line(y="price", use_index=True)
```

```python
price_data["adj_price"] = price_data["price"] / price_data["cpi"] *100

price_data.plot.line(y="adj_price", use_index=True)
```

```python
price_data.dropna()
```

```python
price_data["next_quarter"] = price_data["adj_price"].shift(-13)
price_data["change"] = (price_data["next_quarter"] > price_data["adj_price"]).astype(int)
```

```python
price_data = price_data.iloc[2:]
price_data.head(50)
```

```python
price_data["change"].value_counts()
```

```python
predictors = ["interest", "vacancy", "adj_price"]
target = "change"

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np

START = 260     # 5 years
STEP = 52       # 1 year

def predict(train, test, predictors, target):
    rf = RandomForestClassifier(min_samples_split=10, random_state=1)
    rf.fit(train[predictors], train[target])
    preds = rf.predict(test[predictors])
    return preds

def backtest(data, predictors, target):
    all_preds = []
    for i in range(START, data.shape[0], STEP):
        train = price_data.iloc[:i]
        test = price_data.iloc[i:(i+STEP)]
        all_preds.append(predict(train, test, predictors, target))
    
    preds = np.concatenate(all_preds)
    return preds, accuracy_score(data.iloc[START:][target], preds)
```

```python
preds, accuracy = backtest(price_data, predictors, target)
```

```python
accuracy
```

```python
yearly = price_data.rolling(52, min_periods=1).mean()
yearly_ratios = [p + "_year" for p in predictors]
price_data[yearly_ratios] = price_data[predictors] / yearly[predictors]
```

```python
price_data
```

```python
price_data
```

```python
preds, accuracy = backtest(price_data, predictors + yearly_ratios, target)
```

```python
accuracy
```

```python
pred_match = (preds == price_data[target].iloc[START:])
pred_match[pred_match == True] = "green"
pred_match[pred_match == False] = "red"
plot_data = price_data.iloc[START:].copy()

plot_data.reset_index().plot.scatter(x="index", y="adj_price", color=pred_match)
```
