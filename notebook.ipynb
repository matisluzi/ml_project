```python
import pandas as pd
from datetime import timedelta
import ipywidgets as widgets

# Federal Reserve CSV files
fed_files = ["data/MORTGAGE30US.csv", "data/RRVRUSQ156N.csv", "data/CPIAUCSL.csv"]

# Read and concatenate Federal Reserve data
dfs_fed = [pd.read_csv(f, parse_dates=True, index_col=0) for f in fed_files]
fed_data = pd.concat(dfs_fed, axis=1)
fed_data.ffill(inplace = True)
```

```python
# Zillow CSV files
zillow_files = [
    "data/Metro_zhvi_bdrmcnt_1_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv",
    "data/Metro_zhvi_bdrmcnt_2_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv",
    "data/Metro_zhvi_bdrmcnt_3_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv",
    "data/Metro_zhvi_bdrmcnt_4_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv",
    "data/Metro_zhvi_bdrmcnt_5_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv"
]

# Read Zillow CSV files into a list of DataFrames
dfs_zillow = [pd.read_csv(f, parse_dates=True, index_col=0) for f in zillow_files]
dfs_zillow[0]
```

```python
# Widgets for user input

# Bedrooms widget
bedrooms_input = widgets.IntSlider(
    value = 1,
    min = 1,
    max = 5,
    step = 1,
    description = "Bedrooms:"
)
display(bedrooms_input)

# Get location names from Zillow data
location_names = []
for region in dfs_zillow[bedrooms_input.value]['RegionName']:
    location_names.append(region)

# Location widget
location_input = widgets.Dropdown(
    options = location_names,
    description = "Location:"
)
display(location_input)

# Year widget
year_input = widgets.IntSlider(
    value = 2024,
    min = 2024,
    max = 2100,
    step = 1,
    description = "Future Year:"
)
display(year_input)
```

```python
BEDROOM_INPUT = bedrooms_input.value
LOCATION_INPUT = location_input.value
FUTURE_YEAR = year_input.value

# Debugging: Print BEDROOM_INPUT and LOCATION_INPUT
print("BEDROOM_INPUT:", BEDROOM_INPUT)
print("LOCATION_INPUT:", LOCATION_INPUT)
print("FUTURE_YEAR:", FUTURE_YEAR)

# convert location name to index
LOCATION_INPUT = location_names.index(location_input.value)

# Ensure BEDROOM_INPUT and LOCATION_INPUT are within valid range
if 1 <= BEDROOM_INPUT < len(dfs_zillow) and 0 <= LOCATION_INPUT < len(dfs_zillow[BEDROOM_INPUT]):
    zillowFileData = dfs_zillow[BEDROOM_INPUT-1].iloc[LOCATION_INPUT, 5:]  # Extract relevant data

    # Create DataFrame from extracted data
    zillow_data_df = pd.DataFrame(zillowFileData)

    # Optional: Clean up DataFrame (e.g., handle missing values)
    zillow_data_df = zillow_data_df.dropna()  # Drop rows with NaN values if needed

    # Now you can work with zillow_data_df as your processed Zillow data
    print(zillow_data_df.head())  # Example: Display first few rows of processed data
else:
    print("Invalid BEDROOM_INPUT or LOCATION_INPUT. Check input values.")
zillow_data_df.index = pd.to_datetime(zillow_data_df.index)
zillow_data_df["month"] =  zillow_data_df.index.to_period("M")
```

```python
price_data["next_quarter"] = price_data["adj_price"].shift(-13)
price_data["change"] = (price_data["next_quarter"] > price_data["adj_price"]).astype(int)
```

```python
predictors = ["interest", "vacancy", "adj_price"]
target = "change"

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np

START = 60     # 5 years
STEP = 12       # 1 year

def predict(train, test, predictors, target):
    rf = RandomForestClassifier(min_samples_split=10, random_state=1)
    rf.fit(train[predictors], train[target])
    preds = rf.predict(test[predictors])
    return preds

def backtest(data, predictors, target):
    all_preds = []
    for i in range(START, data.shape[0], STEP):
        train = price_data.iloc[:i]
        test = price_data.iloc[i:(i+STEP)]
        all_preds.append(predict(train, test, predictors, target))
    
    preds = np.concatenate(all_preds)
    return preds, accuracy_score(data.iloc[START:][target], preds)
```

```python
preds, accuracy = backtest(price_data, predictors, target)
preds
```

```python
accuracy
```

```python
pred_match = (preds == price_data[target].iloc[START:])
pred_match[pred_match == True] = "green"
pred_match[pred_match == False] = "red"
plot_data = price_data.iloc[START:].copy()

plot_data.reset_index().plot.scatter(x="index", y="adj_price", color=pred_match)
```

```python
yearly = price_data.rolling(52, min_periods=1).mean()
yearly_ratios = [p + "_year" for p in predictors]
price_data[yearly_ratios] = price_data[predictors] / yearly[predictors]
```

```python
def estimate_future_adj_price(data, future_year):
    # Calculate monthly changes in adj_price
    monthly_changes = data['adj_price'].diff()
    
    # Calculate annualized growth rates based on monthly changes
    # Group monthly changes by year and calculate total change for each year
    yearly_changes = monthly_changes.groupby(data.index.year).sum()
    
    # Convert yearly changes to percentage annual growth rates
    yearly_growth_rates = (yearly_changes / data['adj_price'].groupby(data.index.year).first()).mean()
    
    # Calculate average annual growth rate based on yearly growth rates
    average_growth_rate = yearly_growth_rates.mean()
    print("Average Annual Growth Rate based on Monthly Changes:", average_growth_rate)
    
    # Prepare a new data point for the future year with average rates
    future_data_point = {
        'interest': data['interest'].mean(),   # Use average interest rate
        'vacancy': data['vacancy'].mean(),     # Use average vacancy rate
        'adj_price': data['adj_price'].iloc[-1]  # Use the last known adj_price value from your dataset
    }
    print(future_data_point['adj_price'])
    # Convert the data point into a DataFrame
    future_df = pd.DataFrame([future_data_point])
    
    # Perform backtesting to get predictions using the trained model
    predictors = ["interest", "vacancy", "adj_price"]
    target = "change"
    # Assuming 'backtest' function is defined correctly
    preds, accuracy = backtest(data, predictors, target)
    
    # Use the predicted change to estimate the future adj_price
    predicted_change = preds[-1]  # Get the last predicted change (for the last test period)
    last_known_adj_price = future_data_point['adj_price']
    
    # Calculate the number of years into the future
    years_into_future = future_year - data.index[-1].year
    
    # Apply compound growth rate to estimate future adj_price
    future_adj_price = last_known_adj_price * (1 + average_growth_rate)**years_into_future
    
    if predicted_change == 0:
      reduction_factor = (1 - average_growth_rate)
      future_adj_price = last_known_adj_price * reduction_factor

    return future_adj_price

# Example usage:
# Assuming 'price_data' is your DataFrame containing historical data
future_adj_price = estimate_future_adj_price(price_data, FUTURE_YEAR)
print(f"Estimated adj_price for year {FUTURE_YEAR}: ${future_adj_price:.2f}")
```
